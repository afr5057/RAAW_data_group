```python
healthcare_tickers = df.loc[df["Sector"] == "Health Care"]
healthcare_tickers.head()
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.EcoParameters_db
collection = db.StockPrice
```

```python
tickers = list(healthcare_tickers["Symbol"])
for ticker in tickers:
    data = yf.download(ticker, start="2014-05-03", end="2019-05-03")
```

```python

data = yf.download("ALC", start="2014-05-03", end="2019-05-03")
data.reset_index(inplace=True)
data["Ticker"] = "ALC"
data = data[["Ticker","Date","Close","Volume"]]
list_dict_prices_stock = []
for item, row in data.iterrows():
    
    dict_row ={
        "Ticker" : row["Ticker"],
        "Date" : row["Date"].strftime('%Y-%m-%d'),
        "Close" : row["Close"],
        "Volume" : row["Volume"]
    }
    list_dict_prices_stock.append(dict_row)

db.StockPrice.insert_many(list_dict_prices_stock)
```
