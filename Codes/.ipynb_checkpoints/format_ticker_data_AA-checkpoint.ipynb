```python
tickers = ["AAPL", "MSFT", "GOOGL", "GOOG", "FB", "CSCO", "INTC", "TSM", "ORCL", "SAP", "ADBE", "AVGO", "CRM", "IBM", "ACN", "NVDA", "TXN", "QCOM", "ASML", "VMW", "INTU", "SNE", "BIDU", "NOW", "DELL", "INFY", "MU", "WDAY", "ADI", "AMAT", "CTSH", "ADSK", "ATVI", "NTES", "ERIC", "RHT", "TEL", "SQ", "NXPI", "LRCX", "HPQ", "TWTR", "APH", "XLNX", "NOK", "WIT", "AMD", "FTV", "EA", "MSI"] 
```

```python
start_date = "2010-01-01"
end_date = dt.date.today()
```

```python
data = yf.download(tickers, start=start_date, end=end_date)
```

```python
data.head()
```

```python
data.reset_index(inplace=True)
```

```python
data.head()
```

```python
#data['Quarter'] = 'Q' + test_open["Quarter"].map(str) +'-'+ test_open["Year"].map(str)

data['Quarter'] = 'Q' + ((data["Date"].dt.month-1)//3+1).map(str) +'-'+ (data["Date"].dt.year).map(str)
data = data.sort_values('Date')
```

```python
data.head()
```

```python
# data.to_csv('C:\\Users\\alaq3\\data.csv')
```

```python
data_quaterly = data.groupby(data['Quarter'], sort=False).mean()
```

```python
data_quaterly.head()
```

```python

open_df = data_quaterly[['Open']]
close_df = data_quaterly[['Close']]
volume_df = data_quaterly[['Volume']]
```
